Step 1: Import Required Libraries

```python
import pandas as pd
```

Step 2: Load the Dataset

```python
df = pd.read_excel('/content/IMDB_1000.xlsx')   # change file name if needed
print(df.head())
```

Step 3: Understand Dataset (Rows, Columns, Data Types)

```python
print("Dataset Shape:", df.shape)
print(df.info())
```

Step 4: Check Missing Values

```python
print("Missing Values:\n", df.isnull().sum())
```

Step 5: Data Cleaning (Remove Duplicates)

```python
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

Step 6: Handle Missing Values

```python
# Fill missing rating with average rating
df['IMDB_Rating'] = df['IMDB_Rating'].fillna(df['IMDB_Rating'].mean())

# Fill missing votes with 0
df['No_of_Votes'] = df['No_of_Votes'].fillna(0)

# Convert votes column to integer
df['No_of_Votes'] = df['No_of_Votes'].astype(int)
```

Step 7: Basic Statistics

```python
print("Average Rating:", round(df['IMDB_Rating'].mean(), 2))
print("Highest Rating:", df['IMDB_Rating'].max())
print("Lowest Rating:", df['IMDB_Rating'].min())
print("Total Movies:", df['Series_Title'].nunique())
```

Step 8: Top 10 Highest Rated Movies

```python
top_rated_movies = df.sort_values(by="IMDB_Rating", ascending=False).head(10)
print("\nTop 10 Highest Rated Movies:\n")
print(top_rated_movies[['Series_Title', 'Released_Year', 'IMDB_Rating', 'No_of_Votes']])
```

Step 9: Top 10 Most Voted Movies

```python
most_voted_movies = df.sort_values(by="No_of_Votes", ascending=False).head(10)
print("\nTop 10 Most Voted Movies:\n")
print(most_voted_movies[['Series_Title', 'Released_Year', 'IMDB_Rating', 'No_of_Votes']])
```

Step 10: Movies Released After 2015

```python
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
movies_after_2015 = df[df['Released_Year'] > 2015]
print("\nMovies Released After 2015:\n")
print(movies_after_2015[['Series_Title', 'Released_Year', 'IMDB_Rating']].head(10))
```

Step 11: Movies Count Per Year

```python
movies_per_year = df['Released_Year'].value_counts().sort_index()
print("\nMovies Count Per Year:\n")
print(movies_per_year)
```

Step 12: Average Rating Per Year

```python
avg_rating_per_year = df.groupby("Released_Year")['IMDB_Rating'].mean().sort_index()
print("\nAverage Rating Per Year:\n")
print(avg_rating_per_year)
```

Step 13: Genre Analysis (Top Genres)

```python
df_genre = df.copy()

# Split genres into list
df_genre['Genre'] = df_genre['Genre'].str.split(",")

# Convert list into rows
df_genre = df_genre.explode("Genre")

# Remove extra spaces
df_genre['Genre'] = df_genre['Genre'].str.strip()

# Top 10 genres by movie count
top_genre_count = df_genre['Genre'].value_counts().head(10)
print("\nTop 10 Genres by Movie Count:\n")
print(top_genre_count)

# Top 5 genres by average rating
top_genre_rating = df_genre.groupby("Genre")['IMDB_Rating'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 Genres by Average Rating:\n")
print(top_genre_rating)
```

Step 14: Find Movies with Rating Above 8.5

```python
high_rated_movies = df[df['IMDB_Rating'] > 8.5]
print("\nMovies with Rating above 8.5:\n")
print(high_rated_movies[['Series_Title', 'Released_Year', 'IMDB_Rating', 'No_of_Votes']])
```

Step 15: Save Output Files (Project Result)

```python
top_rated_movies.to_csv("Top_Rated_Movies.csv", index=False)
most_voted_movies.to_csv("Most_Voted_Movies.csv", index=False)
high_rated_movies.to_csv("High_Rated_Movies.csv", index=False)

print("\nCSV Files Saved Successfully!")
```
